```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from google.colab import drive
import json
import os
```

```python
# Function to load CSV file into a DataFrame
def load_dataset(file_path):
    return pd.read_csv(file_path)
```

```python
# Function to load json file into a dataset
def load_json(file_path):
  with open(file_path, "r") as file:
    return json.load(file)
```

```python
# Function to extract subset of data based on start and end year
def extract_subset_data(df, start_year, end_year):
    return df[(df['a'] >= start_year) & (df['a'] <= end_year)]
```

```python
# Function to fit linear regression model
def fit_linear_regression_model(x, y):
    model = LinearRegression()
    model.fit(x, y)
    return model
```

```python
# Function to plot data and regression line
def plot_data_and_regression(x_values, y_values, regression_line, xlabel, ylabel, title, equation, r_squared):
    plt.figure(figsize=(17, 8))
    plt.plot(x_values, y_values, label='Data', linestyle='--')
    plt.plot(x_values, regression_line, label=equation, color='orange', linestyle='--')
    plt.xlabel(xlabel)
    plt.ylabel(ylabel)
    plt.title(title)
    plt.legend()
    plt.show()
```
